# Dropping NA

```python
import pandas as pd
from numpy import nan as NA
df=pd.DataFrame([[1,6,NA],[1,3,NA],[NA,NA,NA],[NA,4,NA]],columns=list('abc'))
#print(df)
#clean=df.dropna()
#print(clean)
#df.dropna(how="all")
#df.dropna(axis=1)
df.dropna(axis=1,how='all')
```

# Filling NA

```python
import numpy as np
from numpy import nan as NAN
df=pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1]=NA
df.iloc[:2,2]=NA
#display(df.fillna(1.5))
#df.fillna({1:0.5,2:0})
df.fillna(method='ffill',limit=2,axis=1)
```

# Droping duplicates

```python
data=pd.DataFrame({'k1':['one','two']*3+['two'],
                  'k2':[1,1,2,3,3,4,4]})
#print(data.duplicated())
#data=data.drop_duplicates()
#display(data)
data['v1']=['one','two','one','four','one','six','two']
display(data)
data=data.drop_duplicates(['k1','v1'])
display(data)
```

# Renaming Indexes

```python
df=pd.DataFrame([[1,6,NA],[1,3,NA],[NA,4,NA]],columns=['ohio','colorado','new york'],
               index=['ohio','colorado','new york'])
transform=lambda x: x[:4].upper()
df=df.index.map(transform)
print(df)
```

# with dataframe either axis can have Hierarical Index

```python
frame=pd.DataFrame(np.arange(12).reshape((4,3)),
index=[['a','a','b','b'],[1,2,1,2]],
columns=[['ohio','ohio','colorado'],
        ['Green','Red','Green']])
frame.index.names=['key1','key2']
frame.columns.names=['state','color']
display(frame)
display(frame['ohio'])
```

```python
frame.swaplevel('key2','key1')
```

```python
frame.sort_index(level=1)
```

```python
frame.swaplevel(0,1).sort_index(level=0)
```
